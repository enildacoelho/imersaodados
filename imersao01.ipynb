# Imerssão de dados Alura - Dia 01


```python
import pandas as pd
import matplotlib as plt

# Carrega dados (formato  csv, compressão do tipo zip) a partir de uma URL 
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
dados = pd.read_csv(url_dados, compression='zip')

# Inicia uma análise exploratória do conjunto de dados

dados.head()
```

```python
dados['tratamento']
```

```python
# Identifica os valores únicos para tratamento
dados['tratamento'].unique() 
```

```python
# Identifica quantos tratamentos
dados['tratamento'].value_counts()
```

```python
# Identifica os valores de tempo
dados['tempo'].unique
```

```python
dados['tempo'].value_counts()
dados['tempo'].value_counts().plot.bar()
```

```python
## Identifica variação de dose
dados['dose'].unique()
dados['dose'].value_counts()
```

```python

# Identifica os valores únicos para dose
dados['dose'].unique()
dados['dose'].value_counts()
dados['dose'].value_counts(normalize=True)
dados['dose'].value_counts().plot.bar()
```

```python
# Identifica os valores únicos para droga
dados['droga'].nunique()
```

```python
# Identifica os valores únicos para g_0
dados['g-0'].unique()
```

```python
## Desafio 1: Investigar por que a a classe tratamento é tão desbalanceada? ##
dados['tratamento'].value_counts().plot.bar(figsize=(5,5),title='Tratamentos aplicados')
```

```python
## Desafio 2:  Plottar as 5 últimas linhas da tabelas ##

dados.tail(5)
```

```python
## Desafio 3: Proporção das classes tratamento ##
dados['tratamento'].value_counts()/dados['tratamento'].count()
```

```python
# Desafio 4: Quantos tipos de drogas analisados ##

df_drogas = dados['droga'].unique()
df_drogas.size

# dados['droga'].nunique()
```

```python
# Desafio 5: Procurar na documentação o método query ##

dados.query('dose=="D1"', inplace = False)
```

```python
# Desafio 6: Colocar titulo no gráfico ##
dados['tratamento'].value_counts().plot.bar(figsize=(5,5),title='Tratamentos aplicados')

```
