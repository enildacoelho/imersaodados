```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
# Carrega dados (formato  csv, compressão do tipo zip) a partir de uma URL 
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
dados = pd.read_csv(url_dados, compression='zip')

# Inicia uma análise exploratória do conjunto de dados

dados.head()
```

```python

## Customizando o Gráfico 

fig, ax= plt.subplots()
cores={'blue', 'yellow'}

x=dados['tratamento'].value_counts()
ax.pie(x, colors=cores, autopct='%.0f%%',explode=(0, 0.1), shadow=True )
ax.set_title('Tratamentos')
```

```python
## Customizando o Gráfico com Matplotlib
fig, ax= plt.subplots()

x=dados['tratamento'].unique()
y=dados['tratamento'].value_counts()
ax.bar(x,y)
ax.set_facecolor('green')

ax.set_title('Tratamentos com_controle e com_droga')
```
