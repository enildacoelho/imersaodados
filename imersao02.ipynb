````python
## Desafio 1. Ordenar o gráfico countplot

cod_compostos = dados['composto'].value_counts().index[0:5]
cod_compostos
plt.figure(figsize=(8,9))
sns.countplot(x='composto',data=dados.query('composto in @cod_compostos'), order=cod_compostos)
````

````python
## Desafio 2. Melhorar a visualização aumentando o tamanho da fonte
ax= sns.countplot(x='composto',data=dados.query('composto in @cod_compostos'), order=cod_compostos)
ax.set_title('Top 5 compostos usados no experimento', fontdict={'fontsize': 15, 'fontweight':'bold'})
ax.set_xlabel('Compostos',fontsize = 12)
ax.set_ylabel('Quantidade',fontsize = 12)
plt.figure(figsize=(8,9))
plt.show();
````

**"Mean"**
**Média**
A média é uma medida de tendência central que indica o valor onde estão concentrados os dados de um conjunto de valores, 


**STD()**
**neDesvio Padrão**
O desvio padrão também é uma medida de dispersão, que indica quanto os dados estão afastados da média. Um valor de desvio padrão alto indica que os valores estão mais espalhados, mais longe da média, e um desvio padrão baixo indica que os valores estão mais próximos da média. No Pandas fazemos o cálculo através da função std()



````python
### describe "Mean" Média - A média é uma medida de tendência central que indica o valor onde estão concentrados os dados de um conjunto de valores, 
Moda
A moda é simples. Nada mais é que o valor que mais se repete dentro de um conjunto. No Pandas, a moda é calculada através da função mode().
````
