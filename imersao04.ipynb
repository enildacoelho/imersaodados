````python
#quantidade de mecanismos que foram ativos nmoa 
dados_resultados['nmoa']=dados_resultados.drop('id', axis=1).sum(axis=1)
````

````python
#Indica se algum mecanismo foi ativado 

dados_resultados['ativo_moa']=dados_resultados['nmoa']!=0
````

````python
# Identifica os elementos (columns) para cada grupo 'inhibitor', 'antagonist', 'agonist', etc. cpmsoderamdp p sufixo col.endswith _sufixo
inibidores = [col for col in dados_resultados if col.endswith('_inhibitor')]
antagonista = [col for col in dados_resultados if col.endswith('_antagonist')]
agonista = [col for col in dados_resultados if col.endswith('_agonist')]
ativador = [col for col in dados_resultados if col.endswith('_activator')]
dados_resultados['nmoa_inibidor']=dados_resultados.loc[:,inibidores].sum(axis=1)
dados_resultados['nmoa_antagonista']=dados_resultados.loc[:,antagonista].sum(axis=1)
dados_resultados['nmoa_agonista']=dados_resultados.loc[:,agonista].sum(axis=1)
dados_resultados['nmoa_ativador']=dados_resultados.loc[:,ativador].sum(axis=1)
````

````python
#faz um merge através do id

dados_merge= pd.merge(dados, dados_resultados[['id','nmoa','ativo_moa','nmoa_inibidor','nmoa_antagonista','nmoa_agonista','nmoa_ativador']], on='id')
dados_merge.head()
````
